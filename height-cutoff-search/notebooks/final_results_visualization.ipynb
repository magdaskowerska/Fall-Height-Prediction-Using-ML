# Fall Height Search Cutoff Experiments: Model Performance Comparison

## Import Libraries

```python
import sys
sys.path.append('..')

from scripts.utils import display_mae_scores_lineplot
import pandas as pd
```

## Loading data

```python
df_height_capped_without_outliers_thresh3=pd.read_csv('../results/kernel_ridge_height_capped_outliers_removed_thresh3/mae_scores_results.csv')
df_height_capped_without_outliers_thresh1_5=pd.read_csv('../results/kernel_ridge_height_capped_outliers_removed_thresh1_5/mae_scores_results.csv')
df_height_capped_with_outliers=pd.read_csv('../results/kernel_ridge_height_capped_with_outliers/mae_scores_results.csv')
df_height_removed_with_outliers=pd.read_csv('../results/kernel_ridge_height_removed_with_outliers/mae_scores_results.csv')
```

```python
df_height_capped_without_outliers_thresh3=df_height_capped_without_outliers_thresh3[df_height_capped_without_outliers_thresh3["model_name"]=="Kernel Ridge Polynomial Kernel"]
df_height_capped_without_outliers_thresh1_5=df_height_capped_without_outliers_thresh1_5[df_height_capped_without_outliers_thresh1_5["model_name"]=="Kernel Ridge Polynomial Kernel"]
df_height_capped_with_outliers=df_height_capped_with_outliers[df_height_capped_with_outliers["model_name"]=="Kernel Ridge Polynomial Kernel"]
df_height_removed_with_outliers=df_height_removed_with_outliers[df_height_removed_with_outliers["model_name"]=="Kernel Ridge - Polynomial Kernel"]
```

```python
df_height_capped_without_outliers_thresh3['model_name']='Kernel Ridge - Polynomial Kernel - SVD\n on Dataset with Height Capped\n Without Outliers (Threshold 3.0)'
df_height_capped_without_outliers_thresh1_5['model_name']='Kernel Ridge - Polynomial Kernel - SVD\n on Dataset with Height Capped\n Without Outliers (Threshold 1.5)'
df_height_capped_with_outliers['model_name']='Kernel Ridge - Polynomial Kernel - SVD\n on Dataset with Height Capped\n With Outliers'
df_height_removed_with_outliers['model_name']='Kernel Ridge - Polynomial Kernel - SVD\n on Dataset with Records Over\n Height Limit Removed'
```

```python
all_results = pd.concat([
    df_height_capped_without_outliers_thresh3,
    df_height_capped_without_outliers_thresh1_5,
    df_height_capped_with_outliers, 
    df_height_removed_with_outliers
], ignore_index=True)
```

## Visualize MAE results as line plot 

```python
plot_path=display_mae_scores_lineplot("../results", all_results,std_flag=False)
print(f"Plot saved at {plot_path}")
```
