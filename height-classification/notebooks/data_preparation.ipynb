## Data Preparation for Height Classification Experiments

```python
import pandas as pd
```

```python
df_berlino=pd.read_csv('../data/dati_berlino_cleaned.csv')
df_milano_train=pd.read_csv('../data/dati_milano_train_cleaned.csv')
df_milano_test=pd.read_csv('../data/dati_milano_test_cleaned.csv')
```

```python
all_data=pd.concat(
    [
        df_berlino,
        df_milano_train,
        df_milano_test
    ], ignore_index=True
)
```

```python
all_data.to_csv('../data/berlin_milan_data_combined.csv', index=False)
```
